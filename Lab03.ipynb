```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
# Load and preprocess the data
data = np.genfromtxt("lab01_data1.csv", delimiter=',', skip_header=1)
#print(data)
```

```python
#normalize the data
#data = (data - np.mean(data)) / np.std(data)
#print(data)
```

```python
# convert to df with columns population and profit
#df = pd.DataFrame(data, columns=['Population', 'Profit'])

# normalization
#df_norm = (df - df.mean()) / df.std()
```

```python
# Initialize variables
theta0 = 0
theta1 = 0
alpha = 0.01
num_iterations = 1500
# Lists to store cost function values for plotting
cost_history = []
# Gradient Descent optimization
for i in range(num_iterations):
  # Calculate predictions and errors
  x = data[:, 0] # City Population
  #print(x)
  y = data[:, 1] # Food Truck Profit
  predictions = theta0 + theta1 * x
  #print(predictions)
  errors = predictions - y
  # Update parameters
  theta0 -= alpha * (1/len(data)) * sum(errors)
  theta1 -= alpha * (1/len(data)) * sum(errors * x)
  # Calculate cost function and store in history
  cost = (1/(2 * len(data))) * sum(errors ** 2)
  #print(cost)
  cost_history.append(cost)
  #print(cost_history)

# Plotting the loss function over iterations
plt.plot(range(num_iterations), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost Function (J(theta))')
plt.title('Loss Function over Iterations')
plt.show()
# Visualize the optimized model
plt.scatter(x, y, marker='x', color='red')
plt.plot(x, theta0 + theta1 * x, color='blue')
plt.xlabel('City Population')
plt.ylabel('Food Truck Profit')
plt.title('Linear Regression Model')
plt.show()
# Continue with analysis, reflection, and finalizing the code
```
