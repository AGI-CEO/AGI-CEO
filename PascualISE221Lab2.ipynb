```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data
data1 = np.loadtxt("lab01_data1.csv", delimiter=',', skiprows=1)
data2 = np.loadtxt("lab01_data2.csv", delimiter=',', skiprows=1)

# Step 2: Data Cleaning
# I did this manually
```

```python

# Step 3: Data Visualization
# For data1
plt.figure(figsize=(10, 6))
plt.scatter(data1[:, 0], data1[:, 1], marker='x', color='red')
plt.xlabel("City Population (in 10,000)")
plt.ylabel("Food Truck Profit (in $10,000)")
plt.title("City Population vs. Food Truck Profit")
plt.show()

# Convert data1 to a Pandas DataFrame with column names
data1_df = pd.DataFrame(data1, columns=["City Population (in 10,000)", "Food Truck Profit (in $10,000)"])

# normalization
data1_df_norm = (data1_df-data1_df.min())/(data1_df.max()-data1_df.min())

print(data1_df.describe())
print(data1_df_norm.describe())

# normalization plot
plt.figure(figsize=(10, 6))
plt.scatter(data1_df_norm["City Population (in 10,000)"], data1_df_norm["Food Truck Profit (in $10,000)"], marker='x', color='red')
plt.xlabel("City Population (in 10,000)")
plt.ylabel("Food Truck Profit (in $10,000)")
plt.title("Normalization- City Population vs. Food Truck Profit")
plt.show()

# mean normalization
data1_df_mean = (data1_df-data1_df.mean())/(data1_df.max()-data1_df.min())
print(data1_df_mean.describe())

# mean normalization plot
plt.figure(figsize=(10, 6))
plt.scatter(data1_df_mean["City Population (in 10,000)"], data1_df_mean["Food Truck Profit (in $10,000)"], marker='x', color='red')
plt.xlabel("City Population (in 10,000)")
plt.ylabel("Food Truck Profit (in $10,000)")
plt.title("Mean Normalization - City Population vs. Food Truck Profit")
plt.show()

# standardization
data1_df_std = (data1_df-data1_df.mean())/data1_df.std()
print(data1_df_std.describe())

# standardization plot
plt.figure(figsize=(10, 6))
plt.scatter(data1_df_std["City Population (in 10,000)"], data1_df_std["Food Truck Profit (in $10,000)"], marker='x', color='red')
plt.xlabel("City Population (in 10,000)")
plt.ylabel("Food Truck Profit (in $10,000)")
plt.title("Standardization - City Population vs. Food Truck Profit")
plt.show()


```

# Reflection:

Normalization, mean normalization, and standardization are three common ways to scale data so that different features become easier to compare. Normalization changes all the values so they fall between 0 and 1, which is simple and straightforward, but it can still be thrown off by very large or very small values. Mean normalization shifts each feature so that its average is zero by subtracting the mean and dividing by the range, which helps reduce some types of bias. Standardization makes each feature have an average of zero and a standard deviation of one, which is particularly helpful for models that rely on normal distributions or distance measures. However, it doesn’t limit values to a specific range. The best method depends on the data, the type of model being used, and how much outliers affect the results.
